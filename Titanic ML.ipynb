# Reading the data

```python
import numpy as np
import pandas as pd
df_train=pd.read_csv('E:/Kaggle Competitions/Titanic/train.csv')
df_test=pd.read_csv('E:/Kaggle Competitions/Titanic/train.csv')
```

# Data preprocessing

```python
df_train.drop(['Name'],axis=1,inplace=True)
```

```python
df_train.drop(['Cabin'],axis=1,inplace=True)
```

```python
df_train.drop(['Ticket'],axis=1,inplace=True)
```

```python
sex={'male':0,'female':1}
embark={'S':0,'C':1,'Q':2,None:1}
df_train.replace({'Sex':sex},inplace=True)
df_train.replace({'Embarked':embark},inplace=True)
```

```python
mean=df_train['Age'].mean()
age={None:mean}
df_train.replace({'Age':age},inplace=True)
```

```python
df_train['Age']=df_train['Age'].astype(int)
df_train['Fare']=df_train['Fare'].astype(int)
df_train['Embarked']=df_train['Embarked'].astype(int)
```

```python
df_train
```

# Train test split

```python
from sklearn.model_selection import train_test_split
x=df_train[['PassengerId','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y=df_train[['Survived']]
X_train,X_test,y_train,y_test= train_test_split(x,y,random_state=0)
```

# K-nn

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=12)
knn.fit(X_train,y_train)
train_score=knn.score(X_train,y_train)
test_score=knn.score(X_test,y_test)
train_score,test_score
```
